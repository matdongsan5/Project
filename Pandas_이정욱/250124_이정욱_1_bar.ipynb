[전처리 목표]
- 남북한 간 동일분류만 비교
- 결측치가 많은 북한 원자력, 남한 신재생은 제거.
- 수치 타입을 object에서 변경.

```python
## 후에 label에 출력하기 위해 영문명으로 변경 및 컬럼 통합삭제
koreaDF['발전 전력별'] = ['ROK-total', 'ROK-water','ROK-fire','ROK-radi','ROK-new','DPRK-total','DPRK-water','DPRK-fire','DPRK-radi']
koreaDF.drop(columns='전력량 (억㎾h)', inplace=True)
koreaDF.index = koreaDF['발전 전력별']
koreaDF.drop(columns='발전 전력별', inplace=True)
koreaDF
```

```python
## 의미없다고 생각되는 행 삭제.
koreaDF.drop('ROK-new',inplace=True)
koreaDF.drop('DPRK-radi',inplace=True)
koreaDF
```

```python
# 데이터타입 변경
koreaDF.astype(int)
```

```python
koreaDF = koreaDF.T
koreaDF
```

[4] 데이터 시각화 <hr>

```python
len(koreaDF.index[::2]), len(koreaDF['ROK-total'][::2])
```

```python
from matplotlib import rc
from matplotlib import font_manager
#폰트 설정.
font_path = r'C:\Users\matdo\AppData\Local\Microsoft\Windows\Fonts\NanumGothic.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
# print(f"폰트 이름: {font_name}")

# 남북한의 전력 비교 분석
plt.figure(figsize=[12,6])
xData = koreaDF.index[::2]
yData1 = koreaDF['ROK-total'][::2]
yData2 = koreaDF['DPRK-total'][::2]

print(xData.shape, yData1.shape, yData2.shape)

x = np.arange(len(koreaDF.index[::2]))
width = 0.2

for i, value in  enumerate(yData1):
    plt.text(i-width, value + 55, str(value), ha='center') 
for i, value in  enumerate(yData2):
    plt.text(i+width, value + 55, str(value), ha='center') 



plt.bar(x-width, yData1, label='ROK', width=0.4)
plt.bar(x+width, yData2, color='r', label='DPRK', width=0.4)
plt.title(f'ROK-total& DPRK-total' , fontdict={'weight':'bold', 'size':'x-large'})
plt.legend()
plt.xticks(x,koreaDF.index[::2])
plt.ylabel('elec')
plt.xlabel('year')
plt.show()
```
