[도표를 월별에서 분기별로 통합.]

```python
tbDF['1분기'] = [tbDF.loc[i,'1월':'3월'].sum() for i in range(7)]
tbDF['2분기'] = [tbDF.loc[i,'4월':'6월'].sum() for i in range(7)]
tbDF['3분기'] = [tbDF.loc[i,'7월':'9월'].sum() for i in range(7)]
tbDF['4분기'] = [tbDF.loc[i,'10월':'12월'].sum() for i in range(7)]
for i in range(1,13):
    tbDF.drop(f'{i}월', axis=1, inplace=True)
tbDF
```

```python
tbDF['계'] /= 1000
tbDF['1분기'] /= 1000
tbDF['2분기'] /= 1000
tbDF['3분기'] /= 1000
tbDF['4분기'] /= 1000
tbDF['연도'] =['2018', '2019', '2020', '2021', '2022', '2023', '2024']
tbDF.rename(columns={'계': '계(단위: 1m)'}, inplace=True)
tbDF.to_excel('./DATA/관광수지1824.xlsx')
```
